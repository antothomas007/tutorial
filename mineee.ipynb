```python
def Palindrome(a):
    left_pos = 0
    right_pos = len(a) - 1
    
    while right_pos >= left_pos:
        if not a[left_pos] == a[right_pos]:
            return False
        left_pos += 1
        right_pos -= 1
    return True
print(Palindrome('adfgha')) 
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class ShoppingCart:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def get_total_cost(self):
        total_cost = 0.0
        for product in self.products:
            total_cost += product.price
        return total_cost

product1 = Product("Widget", 10.0)
product2 = Product("Gadget", 15.0)

cart = ShoppingCart()
cart.add_product(product1)
cart.add_product(product2)

total_cost = cart.get_total_cost()
print("Total Cost of Products in the Cart:", total_cost)
```

```python
#NumPy Sorting Arrays
#Sorting means putting elements in an ordered sequence.

#Ordered sequence is any sequence that has an order corresponding to elements, like numeric or alphabetical, ascending or descending.

#The NumPy ndarray object has a function called sort(), that will sort a specified array.

import numpy as np

arr = np.array([3, 2, 0, 1])

print(np.sort(arr))
```

```python
##Sorting reverse order(1st do ascending order then do reverse)
array = np.array([7, 5, 3, 2, 6, 1, 4])

# Sort in ascending order
sorted_array = np.sort(array)
# [1, 2, 3, 4, 5, 6, 7]

# Reverse the sorted array
reverse_array = sorted_array[::-1]
# [7, 6, 5, 4, 3, 2, 1]

print(reverse_array)
```

```python
#Creating the Filter Array


import numpy as np

arr = np.array([41, 42, 43, 44])

# Create an empty list
filter_arr = []

# go through each element in arr
for element in arr:
    
  # if the element is higher than 42, set the value to True, otherwise False:
    if element > 42:
        filter_arr.append(True)
    else:
        filter_arr.append(False)

newarr = arr[filter_arr]

print(filter_arr)
print(newarr)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
```

```python
df=pd.read_csv(r'C:\Users\DS\Downloads\Universities.csv')
```

```python
df
```

```python
def norm_fun(i):
    x=(i-i.mean())/(i.std())
    return(x)
```

```python
df_norm=norm_fun(df.iloc[:,1:6])
df_norm
```

```python
z=linkage(df_norm,method='complete',metric='euclidean')
```

```python
plt.figure(figsize=(15,10))
plt.title('clustering')
sch.dendrogram(
                z,
              leaf_rotation=0,
              leaf_font_size=8,
              )
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
h_complete=AgglomerativeClustering(n_clusters=3,linkage='complete',affinity='euclidean').fit(df_norm)
```

```python
h_complete.labels_
```

```python
cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
```

```python
df['cluster']=cluster_labels
df
```

```python
df.groupby(df['cluster']).mean()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
ds=pd.read_csv(r'C:\Users\DS\Downloads\Universities.csv')
```

```python
ds
```

```python
x=ds.iloc[:,1:]
```

```python
x
```

```python
wcss=[]
for i in range(1,15):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=1)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,15),wcss)
plt.title('elbow methos')
plt.xlabel('number of cluster')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=1)
y=kmeans.fit_predict(x)
```

```python
s=pd.Series(y)
s
```

```python
ds['clust']=s
ds
```

```python
ds.iloc[:,[7,0,1,2,3.4,5,6]]
```

```python
ds.groupby(ds.clust).mean()
```

```python
import pandas as pd
from apyori import apriori
```

```python
records = []
for i in range(0, 7501):
    records.append([str(ds.values[i,j]) for j in range(0, 20)])
```

```python
records[0]
```

```python
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2,min_lift=3,min_length=2)
association_results = list(association_rules)
```

```python
print("There are {} Relation derived.".format(len(association_results)))
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
email=pd.read_csv(r'C:\Users\DS\Downloads\emails(4).csv')
```

```python
email
```

```python
email.shape
```

```python
target=email.iloc[:,3001]
```

```python
target
```

```python
features=email.iloc[:,1:3001]
features
```

```python
email.isnull().sum()
```

```python
scalar = StandardScaler()
scalar.fit(features)
scaled_data = scalar.transform(features)
```

```python
scaled_data
```

```python
a=pd.DataFrame(scaled_data)
a
```

```python
pca = PCA(n_components = 5)
pca.fit(a)
x_pca=pca.transform(a)
```

```python
x_pca.shape
```

```python
x_pca
```

```python
x=pd.DataFrame(x_pca)
x
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,target, test_size = 0.1, random_state=1)
```

```python
modeldt=DecisionTreeClassifier(criterion='entropy')
modeldt.fit(X_train,y_train)
preddt=modeldt.predict(X_test)
acc=np.mean(preddt==y_test)
```
